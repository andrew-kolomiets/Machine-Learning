```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
frame = pd.read_csv('data/data04.csv', sep = ';', names =['X1', 'X2', 'Y'] )
print (frame)
```

```python
fig, ax = plt.subplots()
color = ['red' if value == 1 else 'blue' for value in frame.Y]
ax.scatter(frame.X1, frame.X2, marker = 'o', color = color)

fig.set_figwidth(9,6)    
fig.set_figheight(9,6)    
plt.show()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(frame[['X1','X2']].values,     
                                                    frame['Y'].values, 
                                                    train_size=0.8)

W = 3*np.random.random(size =(2,1)) - 2
print (W)
```

```python
#Навчання моделі

teta = 0     #поріг 
epoch = True    #щоб проходили цикл, поки не позбудемося помилок
while epoch:
    number_of_error =0
    for i in range (len(Y_train)):
        sum_of_mult = np.dot(X_train[i], W)
        if sum_of_mult <=teta:
            y=0
        else:
            y=1
        #Переобчислюємо ваги (якщо це потрібно)
        W[0] = W[0]+0.01*(Y_train[i] - y)*X_train[i][0]
        W[1] = W[1]+0.01*(Y_train[i] - y)*X_train[i][1]
        teta = teta - 0.01*(Y_train[i] - y)
        number_of_error += abs(Y_train[i] - y)
    if number_of_error == 0:
        epoch = False
```

```python
#Виводимо отримані ваги W та поріг teta

print('number_of_error: ', number_of_error)
print ('W = ', W)
print ('teta = ', teta)
```

```python
#Передбачення на навчальній вибірці

y_pred = np.dot(X_train, W)
for elem in range(len(y_pred)):
    if y_pred[elem] <=teta:
        y_pred[elem]=0
    else:
        y_pred[elem]=1
```

```python
fig, ax = plt.subplots()
X2 = np.arange (0, 1, 0.01)
X1 = -(X2*W[1] - teta)/W[0]     #x1*w1+x2*w2-teta = 0
ax.plot(X1, X2)
color = ['red' if value == 1 else 'blue' for value in Y_train]
ax.scatter(X_train[:, 0], X_train[:, 1], marker = 'o', color = color)

fig.set_figwidth(9,6)    
fig.set_figheight(9,6)    
plt.show()
```

```python
errors = 0
for i in range (len(Y_train)):
    errors += abs(Y_train[i] - y_pred[i])
print ('Кількість помилок: ', errors) 
```

```python
#Тестування

y_test_pred = np.dot(X_test, W)
for elem in range(len(y_test_pred)):
    if y_test_pred[elem] <=teta:
        y_test_pred[elem]=0
    else:
        y_test_pred[elem]=1
print(y_test_pred.T)
```

```python
#Істинні значення Y

print(Y_test.T)
```

```python
errors = 0
for i in range (len(Y_test)):
    errors += abs(Y_test[i] - y_test_pred[i])
print ('Кількість помилок: ', errors) 
```
